This Jupyter Notebook demonstrates how to create basic data visualizations using **Matplotlib** in Python. We'll cover:
- Line plots
- Bar plots
- Scatter plots
- Combining multiple plots in one figure

Each section includes a simple example with explanations. This is designed for beginners, so feel free to experiment with the code!

## 1. Setup

```python
# Install Matplotlib and NumPy (run once if needed)
# !pip install matplotlib numpy

# Import libraries
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Line Plot

A **line plot** connects data points with lines, great for showing trends (e.g., temperature over time).

The code below creates a simple line plot with sample data, labels, and a grid.

```python
# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Create line plot
plt.plot(x, y, color='green', marker='o', linestyle='-', linewidth=2, markersize=6)

# Add labels and title
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Simple Line Plot')

# Add grid and display
plt.grid(True)
plt.show()
```

**Tips**:
- Change `color` to `'red'` or `'green'` to try different colors.
- Try `marker='s'` for square markers or `linestyle='--'` for a dashed line or `linestyle=':'` for a dotted line.

## 3. Bar Plot

A **bar plot** displays categorical data with bars, ideal for comparisons (e.g., sales by category).

The code below creates a bar plot with sample categories and values.

```python
# Sample data
# Bar Plot Sample
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 30]

# Create bar plot
plt.bar(categories, values, color='green', width=0.4,label='Line')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Plot')

# Display
plt.show()
```

**Tips**:
- Use `color=['red', 'blue', 'green', 'yellow']` to color each bar differently.
- Adjust `width` (e.g., `0.8`) for wider bars.

## 4. Scatter Plot

A **scatter plot** shows individual data points, useful for visualizing relationships (e.g., height vs. weight).

The code below creates a scatter plot with sample data.

```python
# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 1, 3, 5])

# Create scatter plot
plt.scatter(x, y, color='red', s=100, marker='^')

# Add labels and title
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Simple Scatter Plot')

# Add grid and display
plt.grid(True)
plt.show()
```

**Tips**:
- Change `s=50` for smaller points or `marker='*'` or (other options: 'o', 's', '^'). for star-shaped points.
- Add `alpha=0.5` to make overlapping points semi-transparent.

## 5. Combined Plot

You can combine multiple plot types in one figure. The code below shows a line, scatter, and bar plot together.

```python
# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 1, 3, 5])
# Create figure
plt.figure(figsize=(8, 6))
# Line plot
plt.plot(x, y, color='blue', marker='o', label='Line')
# Scatter plot
plt.scatter(x, y, color='red', s=100, label='Scatter')
# Bar plot (offset for visibility)
plt.bar(x - 0.2, y, color='green', width=0.4, label='Bar')
# Add labels, title, and legend
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Combined Line, Scatter, and Bar Plot')
plt.legend()
plt.grid(True)
# Display
plt.show()
```

**Tips**:
- Adjust `figsize=(width, height)` to resize the plot.
- The `label` parameter and `plt.legend()` create a legend to identify plots.
